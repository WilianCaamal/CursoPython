Para la ejecuci√≥n de un entorno virtual es necesario usar el siguiente comando para generar el entorno virtual, cabe recalcar que por preferencia propia hago uso de WindowsTerminal.

## Creacion de entorno

Comando: ```python -m venv NombreEntorno```

![imagen](imagenes/entorno.png)

Una vez generado el entorno virtual ahora es necesario posicionarnos dentro de la carpeta creada con el comando ``` cd ``` 

![](imagenes/directorio.png)

## Activar entorno

Para activar el entorno virtual es necesario ejecutar el script ```Activate.bat```

![](imagenes/active.png)

## Desactivar entorno

Para desactivar el entorno virtual es necesario ejecutar el script ```Deactivate.bat```
